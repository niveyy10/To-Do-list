```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = {
    'Country': ['China', 'India', 'USA', 'Indonesia', 'Pakistan', 'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico'],
    'Population': [1425887337, 1417173173, 339996563, 277534122, 240485658, 216422446, 223804632, 172452611, 144404086, 129875529]
}


df = pd.DataFrame(data)


sns.set(style="whitegrid")


plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Population', data=df, palette='viridis')


plt.title('Top 10 Most Populous Countries (2024)', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=45)


plt.tight_layout()
plt.show()
```
